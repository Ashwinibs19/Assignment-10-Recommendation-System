```python
import pandas as pd
import numpy as np
```

```python
book = pd.read_csv('books.csv', encoding='latin-1')
```

```python
book
```

# Exploratory Data Analysis

```python
book.head()
```

```python
book.rename(columns={"User.ID":"user_id","Book.Title":"book_title","Book.Rating":"book_rating"},inplace=True)
     
```

```python
book.info()
```

```python
book.isnull().sum()
```

```python
book.duplicated().sum()
```

```python
book1=book.iloc[:,1:]
```

```python
book1
```

```python
#number of unique users in the dataset
len(book1.user_id.unique())
```

```python
#number of unique books in the dataset
len(book1.book_title.unique())
```

```python
book1['book_rating'].value_counts().sort_index()
```

```python
book1.book_title.value_counts()
```

```python
user_book1 = book1.pivot_table(index='user_id',columns='book_title',values='book_rating')
user_book1.head()
```

```python
#Impute those NaNs with 0 values
user_book1.fillna(0, inplace=True)
```

```python
user_book1
```

# User-User Similarity

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_book1.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book1.user_id.unique()
user_sim_df.columns = book1.user_id.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```

# Item-Item Similarity

```python
item_book1 = book1.pivot_table(index='book_title',columns='book_title',values='book_rating')
item_book1.head()
```

```python
#Impute those NaNs with 0 values
item_book1.fillna(0, inplace=True)
```

```python
item_book1
```

```python
item_sim = 1 - pairwise_distances( item_book1.values,metric='cosine')
```

```python
item_sim
```

```python
#Store the results in a dataframe
item_sim_df = pd.DataFrame(item_sim)
item_sim_df
```

```python
#Set the index and column names to book title 
item_sim_df.index = book1.book_title.unique()
item_sim_df.columns = book1.book_title.unique()
```

```python
item_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(item_sim, 0)
item_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar items
item_sim_df.idxmax(axis=1)[0:5]
```
